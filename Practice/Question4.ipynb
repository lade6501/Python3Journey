```python
#Importing math module
import math
```

```python
#Defining solution functoin
def mySolution():
    radius = float(input('Enter radius of circle '))
    area = ( math.pi * (radius * radius))
    return area
```

```python
area_of_circle = mySolution()
print(f'Area = {area_of_circle}')
```
